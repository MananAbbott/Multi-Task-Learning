# Import the Embedding model class

```python
from src.embedding_model import EmbeddingModel
```

# Instantiate the Model

```python
model = EmbeddingModel(normalize=True)
```

# Example 1

```python
sent1 = ["The temperature today is 71 degrees Fahrenheit."]

# Get the embedding for the sentence
embedding1 = model.encode(sent1)

print("Sentence:", sent1)
print("Embedding:", embedding1)
print("Embedding shape:", embedding1.shape)
```

# Example 2

```python
sent2 = ["I am wearing blue colored headphones which are from the Japanese brand Sony.",
         "The symbol '&' is called the 'and' symbol and '!' is called exclamation mark."
         ]

# Get the embedding for the sentence
embedding2 = model.encode(sent2)

print("Sentence:", sent2)
print("Embedding:", embedding2)
print("Embedding shape:", embedding2.shape)
```
